```python
import cv2

cap = cv2.VideoCapture(1)
while(1):
    # get a frame
    ret, frame = cap.read()
    # show a frame
    cv2.imshow("capture", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        cv2.imwrite("a.jpeg", frame)
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
im=cv2.imread("a.jpeg")
```

```python
dst=im.copy()
```

```python
rows,cols,channels=im.shape
a=0.5
b=150
for i in range(rows):
    for j in range(cols):
        for c in range(3):
            color=im[i,j][c]*a+b
            if color>255:           # 防止像素值越界（0~255）
                dst[i,j][c]=255
            elif color<0:           # 防止像素值越界（0~255）
                dst[i,j][c]=0
```

```python
plt.imshow(cv2.cvtColor( dst, cv2.COLOR_BGR2RGB))
```
