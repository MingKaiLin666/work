```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
K = np.linspace(-3.9, 3.9, 30)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = y = np.linspace(-2.5, 2.5, 200)

X, Y = np.meshgrid(x, y)
```

```python
ax = plt.gca(projection='3d')
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
ax.set_zlim(-4,4)

for z in K:
    Z = 4*np.cos(X)**2 + 4*np.sin(Y)**2 + (z - 4)
    plt.contour(X, Y, Z+z, [z])
```
